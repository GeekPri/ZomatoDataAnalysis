```python
#consolidating the RATING types to the common ones
raw_df['RATING_TYPE'].replace(to_replace='Excelente' , value='Excellent', inplace=True) 
raw_df['RATING_TYPE'].replace(to_replace=['Veľmi dobré','Bardzo dobrze','Muy Bueno','Velmi dobré'] , value='Very Good', inplace=True)
raw_df['RATING_TYPE'].replace(to_replace=['Skvělá volba','Dobrze','Bueno','Buono','Dobré','Bom','Skvělé'] , value='Good', inplace=True)
raw_df['RATING_TYPE'].replace(to_replace=['Priemer','Média','Çok iyi'] , value='Average', inplace=True) 
raw_df['RATING_TYPE'].replace(to_replace=['Průměr','Promedio','Ortalama','Muito Bom','İyi'] , value='Poor', inplace=True)
raw_df['RATING_TYPE'].replace(to_replace=['Baik','Biasa','Media','Sangat Baik'] , value='Very Poor', inplace=True) 
```

```python

raw_df['REGION'].replace(to_replace='De MallVeera Desai Area|Fun Republic MallVeera Desai Area' , value='Desai', inplace=True) 
raw_df['REGION'].replace(to_replace=r'Hotel EmeraldJuhu|Juhu Residency Boutique HotelJuhu', 
                         value='Juhu', 
                         inplace=True, 
                         regex=True)
raw_df['REGION'].replace(to_replace='Cooling TowerGoregaon East|Royal Challenge ComplexGoregaon' , value='Goregaon', inplace=True) 
raw_df['REGION'].replace(to_replace='Hotel Highway ViewSanpada' , value='Sandapa', inplace=True) 
raw_df['REGION'].replace(to_replace='De MallVeera Desai Area' , value='Desai', inplace=True) 
raw_df['REGION'].replace(to_replace='Kenilworth MallLinking RoadBandra' , value='Bandra', inplace=True) 
raw_df['REGION'].replace(to_replace='Hotel Royal InnKhar' , value='Innkhar', inplace=True) 
raw_df['REGION'].replace(to_replace='Haiko MallPowai' , value='Powai', inplace=True) 
raw_df['REGION'].replace(to_replace='Hotel Centre PointCBD Belapur' , value='CBD-Belapur', inplace=True) 
raw_df['REGION'].replace(to_replace='Hotel JawaharUlhasnagar|rUlhasnagar' , value='Ulhasnagar', inplace=True) 
raw_df['REGION'].replace(to_replace='Little World MallKharghar' , value='Kharghar', inplace=True) 
raw_df['REGION'].replace(to_replace='Girgaon Chowpatty' , value='Chowpatty', inplace=True) 

raw_df['REGION'].replace(to_replace='K Star MallChembur|Hotel Jewel of ChemburChembur|Hotel Jewel of ChemburChembur' , value='Chembur', inplace=True) 
```

```python
#Q5) Which places have the highest rated restaurant for each Cuisine Type in Mumbai?
highest_rated_df = zomato_df[zomato_df['RATING'] >= 4.6]
```

```python
highest_rated_df

```

```python
fig = px.histogram(highest_rated_df, x = 'REGION', color = 'CUSINE TYPE', title = 'No. of restarants with rating => 4.6 by each places'
                  ) 
fig.update_xaxes(categoryorder="total descending")
fig.update_layout(height=800, 
                  barmode='stack',  # Ensures all colors stack properly
                  legend_title_text='Cuisine Type')
 
fig.update_layout(legend_traceorder="normal") 

fig.show()
```

```python
#  Q6) What is the Avg Price Distibution of highest rated restaurant for each Cuisine Type in Mumbai?
highest_rated_df = zomato_df[zomato_df['RATING'] >= 4.6]
highest_rated_price_df = highest_rated_df.groupby(by = ['REGION', 'CUSINE TYPE'])['PRICE'].mean().reset_index()
```

```python
highest_rated_price_df.head()
```

```python
fig = px.scatter(highest_rated_price_df, x='REGION', y='PRICE', color = 'CUSINE TYPE', symbol = 'CUSINE TYPE',
                title = 'Avg price Distribution of Highly rated restaurants for each cusine types '
                ).update_traces(marker_size = 10)
fig.update_xaxes(categoryorder="total descending")
fig.update_layout(height=800, 
                  barmode='stack',  # Ensures all colors stack properly
                  legend_title_text='Cuisine Type')
 
fig.update_layout(legend_traceorder="normal") 

fig.show() 
```

```python
# Q7) Which areas have a large number of Chinese Restaurant Market?
chinese_df = zomato_df[zomato_df['CUSINE_CATEGORY'].str.contains('Chinese')]
chinese_df
```

```python
chinese_rest_df = chinese_df.groupby(by='REGION').agg({'NAME': 'count', 'PRICE': 'mean'}
                                                     ).rename(columns={'NAME': 'COUNT of Restaurants'}).reset_index()

chinese_rest_df = chinese_rest_df.sort_values('COUNT of Restaurants', ascending=False).head(25)
chinese_rest_df.head()
```

```python
fig = px.bar(chinese_rest_df, x='REGION', y='COUNT of Restaurants',
             color='PRICE', title= 'No. of Chinese Restaurant by places') 
 
fig.show() 
```

```python
#Q8) Is there a relation between Price and Rating by each Cuisine Type?
price_rating_df = zomato_df.groupby(['CUSINE TYPE', 'RATING'])['PRICE'].mean().reset_index()
price_rating_df
```
